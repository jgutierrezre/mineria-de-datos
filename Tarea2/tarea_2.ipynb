Escriba un cuaderno en donde implemente una clase iterable que haga lo siguiente:
1. Al inicio abra el archivo y lo suba a un dataframe de Pandas.
1. Entregue la informaci√≥n de la siguiente cuenta cuando se le pida el siguiente dato.

```python
# implement iterable class that opens a file and saves it to a dataframe and returns each row
class Reader:
    def __init__(self, filename):
        self.df = pd.read_csv(filename)
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.df):
            self.index += 1
            return self.df.iloc[self.index-1]
        else:
            raise StopIteration
```

```python
# read tarea_2.csv with Reader class
reader = Reader('tarea_2.csv')
for row in reader:
    print(row)
    print()
```
